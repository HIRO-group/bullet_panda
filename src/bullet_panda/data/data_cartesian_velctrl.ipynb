```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('robot_data.npy')
```

```python
ee_pos_x = []
ee_pos_y = []
for record in data:
    ee_pos_x.append(record['ee_point'][0])
    ee_pos_y.append(record['ee_point'][1])

plt.plot(ee_pos_x[:len(ee_pos_x)/2], ee_pos_y[:len(ee_pos_y)/2], label='original')
plt.plot(ee_pos_x[len(ee_pos_x)/2:], ee_pos_y[len(ee_pos_y)/2:], label='reverse')
plt.legend()
plt.axes().set_aspect('equal', 'datalim')
plt.xlabel('x')
plt.ylabel('y')
plt.title('end effector trajectory for panda')
plt.show()
```

```python
ee_vel_x = []
ee_vel_y = []
for record in data:
    ee_vel_x.append(record['ee_vel'][0])
    ee_vel_y.append(record['ee_vel'][1])

plt.plot(ee_vel_x[:len(ee_vel_x)/2], label='original')
plt.plot(ee_vel_x[len(ee_vel_x)/2:], label='reverse')

plt.legend()
plt.xlabel('time step')
plt.ylabel('velocity')
plt.title('x-velocity for panda end effector')
plt.show()
```

```python
plt.plot(ee_vel_y[:len(ee_vel_y)/2], label='original')
plt.plot(ee_vel_y[len(ee_vel_y)/2:], label='reverse')

plt.legend(bbox_to_anchor=(1.25, 0.95))
plt.xlabel('time step')
plt.ylabel('velocity')
plt.title('y-velocity for panda end effector')
plt.show()
```
